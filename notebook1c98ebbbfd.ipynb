```python
ds = pd.read_csv("../input/house-prices-advanced-regression-techniques/train.csv")
```

```python
ds
```

```python
classes = {}

for clas in ds:
    classes[clas] = []
    for values in ds[clas]:
        classes[clas].append(values)
        
    
```

```python
cualitative=[]
indexs =[]
for i in classes.keys():
    indexs.append(i)
    
for i in range(len(indexs)):
    try:
        ds[indexs[i]][0] + 1
        
        
    except:
        cualitative.append(i)
```

```python
cualitative_labels = []
start = 0
for i in indexs:
    if start in cualitative:
        cualitative_labels.append(indexs[start])
    start += 1
```

```python
cualitative_labels
```

```python
df = ds[cualitative_labels]
```

```python
df.head()
```

```python
x = pd.get_dummies(df, prefix='Encoded')
y = ds['SalePrice']
```

```python
x
```

```python
ranges_width = 80000
starting_point = 0
discrete_y = {}
a = max(y)/ranges_width
range_number = int(a) +1
for range_number in range(len(y)):
    discretized = []
    for i in y:
        if i in range(starting_point, starting_point+ranges_width):
            discretized.append(i)
            discrete_y[int(range_number)] = discretized
    starting_point += ranges_width
print(discrete_y)
```

```python
indice = 0
for e in y:
    for i in discrete_y.keys():
        for f in discrete_y[i]:
            if e==f:
                y[indice] = i
    indice += 1
            
        
```

```python
x
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(500, input_dim = 234, activation = 'relu'))
model.add(Dense(450, activation = 'tanh'))
model.add(Dense(400, activation = 'tanh'))
model.add(Dense(350, activation = 'tanh'))
model.add(Dense(300, activation = 'tanh'))
model.add(Dense(250, activation = 'tanh'))
model.add(Dense(200, activation = 'tanh'))
model.add(Dense(150, activation = 'tanh'))
model.add(Dense(100, activation = 'tanh'))
model.add(Dense(50, activation = 'tanh'))

model.add(Dense(1, activation = 'softmax'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
for i in range(0, 5):
    model.fit(x, y, epochs=100, batch_size= 64, verbose=1)
```

```python
model.evaluate(x, y)
```
